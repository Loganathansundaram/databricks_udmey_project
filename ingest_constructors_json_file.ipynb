```python
from pyspark.sql.types import IntegerType, StringType, StructField, StructType
from pyspark.sql.functions import col, lit, to_timestamp , current_timestamp

constructors_schema = StructType([StructField('constructorId', IntegerType(), False),
                                 StructField('constructorRef', StringType(), True),
                                 StructField('name', StringType(), True),
                                 StructField('nationality', StringType(), True),
                                 StructField('url', StringType(), True)])
```

```python
construct_df = spark.read.schema(constructors_schema).json(f"{bronze_folder_path}/constructors.json")
# construct_df.display()
```

```python
construct_upd_df = construct_df.withColumnRenamed('constructorId', 'constructor_id') \
                                 .withColumnRenamed('constructorRef', 'constructor_ref') \
                                 .withColumn('ingestion_date', lit(current_timestamp())) \
                                 .drop('url')
```

```python
construct_upd_df.write.mode('overwrite').option('header', True).csv(f"{silver_folder_path}/constructors")
```
